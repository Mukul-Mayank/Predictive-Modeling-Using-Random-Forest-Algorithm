loading libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

dataset

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
df = {
    "x1": x[:, 0],  # First column of x
    "x2": x[:, 1],  # Second column of x
    "y": y          # Labels
}
```

```python
data=pd.DataFrame(df)
```

```python
data.isnull().sum()
```

data visualization

```python
plt.figure(figsize=(20,7))
sns.scatterplot(x="x1",y="x2",data=data,hue="y")
plt.title("Two Moons Dataset")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

model building

```python
x_a=data.iloc[:,:-1]
y_a=data["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,random_state=10,test_size=0.2)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(),n_estimators=30)
```

```python
y_train.shape,y_test.shape
```

```python
x_train.shape,x_test.shape
```

```python
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100
```

```python
sv=SVC()
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=300)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100
```
